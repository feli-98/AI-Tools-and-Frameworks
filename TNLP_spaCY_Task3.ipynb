```python
# Import spaCy and other tools
import spacy
from spacy import displacy

# Load the small English model for NER
nlp = spacy.load("en_core_web_sm")

# Sample Amazon product reviews
reviews = [
    "I love my new Apple iPhone 13! The camera is amazing.",
    "Terrible experience with the Samsung Galaxy. It kept freezing.",
    "The Sony headphones are comfortable and sound great.",
    "Avoid this Lenovo laptop. Battery life is awful.",
    "My new Nike running shoes are super comfy and stylish!"
]

# Define a simple rule-based sentiment analyzer
positive_keywords = ["love", "amazing", "great", "comfortable", "comfy", "stylish"]
negative_keywords = ["terrible", "awful", "avoid", "freezing", "bad"]

# Analyze each review
for review in reviews:
    doc = nlp(review)

    # Extract named entities
    entities = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ in ["PRODUCT", "ORG"]]

    # Rule-based sentiment detection
    review_lower = review.lower()
    sentiment = "Positive" if any(word in review_lower for word in positive_keywords) else \
                "Negative" if any(word in review_lower for word in negative_keywords) else "Neutral"

    # Display results
    print(f"\nğŸ“ Review: {review}")
    print(f"ğŸ” Entities: {entities}")
    print(f"â¤ï¸ Sentiment: {sentiment}")
```
